```python
# Plotting data from a dataframe
import pandas as pd
data = pd.read_csv('gapminder_gdp_oceania.csv', index_col='country')
```

```python
data.loc['Australia'].plot()
plt.xticks(rotation=90)
plt.show()
```

```python
# Transposing and plotting data
data.T.plot()
plt.ylabel('GDP per capita')
plt.xticks(rotation=90)
plt.show()
```

```python
#Use a theme similar to R's ggplot2
plt.style.use('ggplot')
data.T.plot(kind='bar')
plt.ylabel('GDP per capita')
plt.xticks(rotation=90)
plt.show()
```

```python
#Make better x-axis labels by extracting just the year
years=[]
for column in data.columns:
    year = column[-4:]
    years.append(year)
```

```python
print(years)
```

```python
# Turn the Australian data into a list
gdp_australia = data.loc['Australia'].tolist()
print (gdp_australia)
```

```python
# Now plot the gdp data against the years
# Add b- for the correct line style
plt.plot(years, gdp_australia)
plt.show()
```

```python
# Plot both countries together
# Step 1: Accumulator to get the years
years=[]
for column in data.columns:
    year = column[-4:]
    years.append(year)
    
# Step 2: Get two countries worth of data
gdp_australia = data.loc['Australia']
gdp_nz = data.loc['New Zealand']

# Step 3: Now plot them
plt.plot (years, gdp_australia, 'b-', label='Australia')
plt.plot (years, gdp_nz, 'g-', label='New Zealand')

# Step 4: Add a legend
plt.legend(loc='upper left')
plt.xlabel('Year')
plt.ylabel('GDP per capita ($)')
plt.show()





```

```python
# Scatter plots with Matplotlib
plt.scatter (gdp_australia, gdp_nz)
plt.show()
```

```python
# Scatter plot with dataframes
data.T.plot.scatter(x= 'Australia', y = 'New Zealand')
```

```python
data_europe = pd.read_csv('gapminder_gdp_europe.csv', index_col='country')
data_europe.min().plot(label='min')
data_europe.max().plot(label='max')
plt.legend(loc='best')
plt.xticks(rotation=90)
plt.show()
```
